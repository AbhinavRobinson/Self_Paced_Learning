# Data visualisation Challenge

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./../data_sets/movie_metadata/movies_metadata.csv",low_memory=False)
```

```python
title_len = (data["original_title"].apply(lambda x: int(len(x)))).to_numpy()
```

```python
plt.style.use("dark_background")
plt.hist(title_len,bins=12)

plt.ylabel("freq of title")
plt.xlabel("length of title")
plt.title("length of titles")

plt.show()
```
