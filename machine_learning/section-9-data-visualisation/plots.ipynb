# Plots

## line plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### themes

```python
print(plt.style.available)
plt.style.use("seaborn-whitegrid")
```

```python
# dummy data
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1,color='red',label="india",marker='o')
plt.plot(x,y2,color="blue",label="usa",linestyle='dashed',marker="*")
plt.xlabel("time")
plt.ylabel("population")
plt.title("Population With Time")
plt.legend()
plt.show()

```

```python
plt.plot(y1,color='red',label="india",marker='o')
plt.plot(y2,color="blue",label="usa",linestyle='dashed',marker="*")
plt.xlabel("time")
plt.ylabel("population")
plt.title("Population With Time")
plt.legend()
plt.show()
# the index will become the x axis by default
```

## scatter plots

```python
# adjust the figsize of any plot
plt.figure(figsize=(4,4))

plt.scatter(x,y1,color='red',label="india",marker='o')
plt.scatter(x,y2,color="blue",label="usa",linestyle='dashed')
plt.xlabel("time")
plt.ylabel("population")
plt.title("Population With Time")
plt.legend()
plt.show()
```

## bar graph

```python
# changing style
plt.style.use("dark_background")

# defining coordinates
x_coordinates = np.array([0,2,4])

# plotting and comparision
plt.bar(x_coordinates,[10,20,15],width=0.7,label="current year",tick_label=["Gold","Platinum","Silver"],color="yellow")
plt.bar(x_coordinates+0.7,[12,17,6],width=0.7,label="next year",color="orange")
plt.title("metal price comparrision")
plt.xlabel("metal")
plt.ylabel("price")
# scale the hieght of the y graph
plt.ylim(0,30)
plt.legend()
plt.show()
```

## Pie chart

```python
plt.figure(figsize=(4,4))

plt.style.use("dark_background")

subjects = ["Maths","Physics","English","Chem"]
weightage = [20,10,15,5]

plt.title("Marks Percentage")

plt.pie(weightage,labels=subjects,explode=(0,0.1,0,0),autopct='%1.1f%%')
plt.show()
```

## Histogram

```python
plt.figure(figsize=(8,8))

# standard normal distribution
Xsn = np.random.randn(100)
# updating distribution
sigma = 6
u = 60
X1 = Xsn*sigma + u
X2 = Xsn*(sigma+2) + (u/2)

# histogram
plt.hist(X1,alpha=0.9,label="maths")
plt.hist(X2,alpha=0.9,label="physics")

plt.ylabel("freq of students")
plt.xlabel("Marks range")
plt.title("Histogram")
plt.ylim(0,30)
plt.show()
```
