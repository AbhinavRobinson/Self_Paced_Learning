## Seaborn
- built on top of matplotlib
- statistical plotting libraray
- works very well with pandas dataframe object
- <a href="https://seaborn.pydata.org/examples/index.html">Seaborn Gallery</a>

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

## Bar plot

```python
sns.set_style("darkgrid")
# default aggregation is mean
sns.barplot(x="sex",y="total_bill",data=tips)
```

```python
sns.set_style("darkgrid")
# sum of all bills
sns.barplot(x="sex",y="total_bill",data=tips,estimator=np.sum)
```

## Count Plot
- good for categorical visualisation

```python
sns.set_style("darkgrid")
# the aggregation is count of all
sns.countplot(x="sex",data=tips)
```

```python
sns.set_style("darkgrid")
# the aggregation is count of all
sns.countplot(x="time",data=tips)
```

## Box plot
- outliers (points that dont fit in the overall distribution)
- max
- mean/average
- min

```python
sns.set_style("darkgrid")
# they are called quartiles
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
sns.set_style("darkgrid")
# they are called quartiles
sns.boxplot(x="day",y="total_bill",data=tips,hue="smoker")
```

```python
sns.set_style("darkgrid")
# they are called quartiles
sns.boxplot(x="day",y="total_bill",data=tips,hue="sex")
```

## violin plot
- max
- mean
- min

```python
sns.set_style("darkgrid")
# they are called quartiles
sns.violinplot(x="day",y="total_bill",data=tips)
```

```python
sns.set_style("darkgrid")
# they are called quartiles

sns.violinplot(x="day",y="total_bill",data=tips,hue="sex",split=True)
```

## Distribution plot

```python
sns.distplot(tips['total_bill'])
```

```python
# KDE -> kernel density estimation
sns.distplot(tips['total_bill'],kde=False,bins=20)
```

## Kernel Density Estimation plot

```python
sns.kdeplot(tips['total_bill'])
```

## joint distribution plot
- capture relationship between tw0 variables 

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="scatter")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
```

## Pair plot
- capture relation between all the varaibles
- pairwise correaltion

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue="day")
```

## Heatmap
- <a href="https://towardsdatascience.com/why-feature-correlation-matters-a-lot-847e8ba439c4">Correlation</a>
- <a href="https://statisticsbyjim.com/regression/multicollinearity-in-regression-analysis/">Why multilcolinearity effexts a model?</a>
- accepts a correlation matrix

```python
tips.head()
```

```python
# correlation between all non categoral columns
tips_corr = tips.corr()
```

```python
tips_corr
```

```python
sns.heatmap(tips_corr,annot=True)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(tips_corr, vmax=1.0, center=0, fmt='.2f',square=True, linewidths=.7, annot=True, cbar_kws={"shrink": .70})
plt.show();
```

## Pivot table
- relationship between features
- visualising data

```python
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
flights_pivot = flights.pivot_table(index="month",columns="year",values="passengers")
```

```python
flights_pivot
```

```python
plt.subplots(figsize=(8,8))
sns.heatmap(flights_pivot)
```
