```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
eye_cascade = cv2.CascadeClassifier('./eye.xml')
smile_cascade = cv2.CascadeClassifier('./smile.xml')
face_cascade = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
```

```python
print(eye_cascade.empty())
print(smile_cascade.empty())
print(face_cascade.empty())
```

```python
print(eye_cascade.load('eye.xml'))
print(smile_cascade.load('smile.xml'))
print(face_cascade.load('haarcascade_frontalface_default.xml'))
```

```python
frame = cv2.imread("./Train/Jamie_Before.jpg")
frame= cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(frame)
plt.axis("off")
plt.show()
```

```python
smile = smile_cascade.detectMultiScale(frame,1.3,21)
faces = face_cascade.detectMultiScale(frame,1.3,7)
eyes = eye_cascade.detectMultiScale(frame,1.3,3)
```

```python
print(len(smile))
print(len(faces))
print(len(eyes))
```

```python
for (x,y,w,h) in faces:
    offset = 10
    face_Section = frame[y-offset-5:y+h+offset+5,x-offset:x+w+offset]
    cv2.rectangle(frame,(x,y),(x+w,y+h),(255,113,223),3)
    plt.imshow(frame,cmap="gray")
    plt.axis("off")
    plt.show()
```

```python
frame = cv2.imread("./Train/Jamie_Before.jpg")
frame= cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
smile = smile_cascade.detectMultiScale(frame,1.3,21)
faces = face_cascade.detectMultiScale(frame,1.3,7)
for (x,y,w,h) in faces:
    frame = cv2.imread("./Train/Jamie_Before.jpg")
    frame= cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    offset = 10
    face_Section = frame[y-offset-5:y+h+offset+5,x-offset:x+w+offset]
    cv2.rectangle(frame,(x,y),(x+w,y+h),(255,113,223),3)
    frame = frame[y:y+h,x:x+w]
    
    smile = smile_cascade.detectMultiScale(frame,1.3,40)
    # sort smile based on larges area 
    
    eyes = eye_cascade.detectMultiScale(frame,1.3,3)
    # sort eyes based on larges area 
    
    
    frame = cv2.cvtColor(frame, cv2.COLOR_RGB2RGBA)
    
    
    # mouth
    (x2,y2,w2,h2) = smile[0]
    cv2.rectangle(frame,(x2,y2),(x2+w2,y2+h2),(0,0,0),3)
    # (30, 70, 4)
    

    # eye
    (x3,y3,w3,h3) = eyes[0]
    cv2.rectangle(frame,(x3,y3),(x3+w3,y3+h3),(0,0,0),3)
    (x4,y4,w4,h4) = eyes[1]
    cv2.rectangle(frame,(x4,y4),(x4+w4,y4+h4),(0,0,0),3)
    
    plt.imshow(frame)
    plt.axis("off")
    plt.show()
    
    
```
