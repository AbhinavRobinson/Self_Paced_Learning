```python
X = X.values
y = Y.values

# normalise
u = X.mean()
std = X.std()

print("mean",u)
print("std",std)

X = (X-u)/std
```

```python
X = X.reshape((-1,))
y = y.reshape((-1,))
print(X.shape)
print(y.shape)
```

## Section-2 - Linear Regression

```python
def hypothesis(X,theta):
    return theta[0] + np.dot(theta[1],X)
```

```python
def gradient(X,y,theta):
    grad = np.zeros((2,))
    m = X.shape[0]
    grad[0] = np.sum(hypothesis(X,theta)-y)/m
    grad[1] = np.sum(np.dot(hypothesis(X,theta)-y,X))/m
    
    return grad
```

```python
def error(X,y,theta):
    return np.sum((hypothesis(X,theta)-y)**2)/X.shape[0]
```

```python
def gradient_descent(X,y,lr=0.1,max_iter=100):
    theta = np.zeros((2,))
    error_list = []
    theta_list = []
    for i in range(max_iter):
        # compute grad
        grad = gradient(X,y,theta)
        
        # error
        e = error(X,y,theta)
        error_list.append(e)
        
        # update the theta
        theta[0] = theta[0] - lr*grad[0]
        theta[1] = theta[1] - lr*grad[1]
        theta_list.append((theta[0],theta[1]))
    return theta,error_list,theta_list 
```

```python
theta,error_list,theta_list = gradient_descent(X,y) 
```

## Section-4  Scoring model using r2 score

```python
def r2score(y,y_pred):
    num = np.sum((y - y_pred)**2)
    den = np.sum((y - y.mean())**2)
    return (1-(num)/den)*100
```

```python
r2score(y,y_pred)
```
