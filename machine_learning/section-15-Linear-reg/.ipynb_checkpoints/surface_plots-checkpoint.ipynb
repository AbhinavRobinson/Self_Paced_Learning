## Surface plot and  Data visualisation

```python
# mpl_toolkits.mplot3d provides some basic 3D plotting (scatter, surf, line, mesh) tools. 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6,7])

# Return coordinate matrices from coordinate vectors.
# Make N-D coordinate arrays for vectorized evaluations of N-D scalar/vector 
# fields over N-D grids, given one-dimensional coordinate arrays x1, x2,â€¦, xn.

# creating a surface matrix
a,b = np.meshgrid(a,b)

print(a)
print(b)
```

```python
# creating a 3d space
# Figure, which contains all the plot elements.
fig = plt.figure()
# The gca() method figure module of matplotlib library is used to get the current axes.
axes = fig.gca(projection="3d")
plt.show()
```

```python
# creating a 3d space
# Figure, which contains all the plot elements.
fig = plt.figure()
# The gca() method figure module of matplotlib library is used to get the current axes.
axes = fig.gca(projection="3d")
axes.plot_surface(a,b,a+b,cmap="coolwarm")
plt.show()
```

```python
a = np.arange(-1,1,0.02)
b = a

# Return coordinate matrices from coordinate vectors.
# Make N-D coordinate arrays for vectorized evaluations of N-D scalar/vector 
# fields over N-D grids, given one-dimensional coordinate arrays x1, x2,â€¦, xn.

# creating a surface matrix
a,b = np.meshgrid(a,b)

print(a)
print(b)
```

### surface plot

```python
# creating a 3d space
# Figure, which contains all the plot elements.
fig = plt.figure()
# The gca() method figure module of matplotlib library is used to get the current axes.
axes = fig.gca(projection="3d")
axes.plot_surface(a,b,a**2+b**2,cmap="coolwarm")
plt.show()
```

### contour plot

```python
# creating a 3d space
# Figure, which contains all the plot elements.
fig = plt.figure()
# The gca() method figure module of matplotlib library is used to get the current axes.
axes = fig.gca(projection="3d")
axes.contour(a,b,a**2+b**2,cmap="coolwarm")
plt.show()
```
