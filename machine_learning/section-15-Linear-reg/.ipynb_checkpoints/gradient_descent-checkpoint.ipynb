```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(11)
y = (X-5)**2
```

```python
print(X)
print(y)
```

Given a function of f(x) we want to find a value of x which will minimise the the value of f(x)

```python
plt.style.use("seaborn")
plt.plot(X,y)
plt.xlabel("X")
plt.ylabel("f(x)")
plt.show()
```

## Gradient descent

```python
# initialise weights
x = 0
# learing rate
lr = 0.1
# error
error = []

# plot
plt.plot(X,y)
plt.xlabel("X")
plt.ylabel("f(x)")

for i in range(50):
    grad = 2*(x-5)
    x = x - lr*(grad)
    e = (x-5)**2
    error.append(e)
    plt.scatter(x,e,color="red")
```

## plotting error

```python
plt.plot(error)
plt.show()
```
