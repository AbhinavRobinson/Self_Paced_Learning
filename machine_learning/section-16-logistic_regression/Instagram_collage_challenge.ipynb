# Task

You need to resize each pokemon 200X200 before using it. The final image should have size 430 X 430 px

(10 border + 200 pokemon + 10 border + 200 pokemon + 10 border = 430 px) along row and column.

After that you need to convert the final image into a single .csv file of the shape (rows x cols, 3). Your accuracy will be based on percentage of values matched with our file. The CSV file should have a header "r,g,b" as column names. center image is 100*100 with 10 pixel border

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
def collage(img1,img2,img3,img4,img5):
    
    img1 = cv2.resize(img1,(200,200))
    img2 = cv2.resize(img2,(200,200))
    img3 = cv2.resize(img3,(200,200))
    img4 = cv2.resize(img4,(200,200))
    img5 = cv2.resize(img5,(100,100))
    final_image = np.zeros((430,430,3),dtype="int")
    center_image = np.zeros((120,120,3),dtype="int")
    center_image[10:110,10:110,:] = img5
    final_image[10:210,10:210,:] = img1 
    final_image[10:210,220:420,:] = img2
    final_image[220:420,10:210,:] = img3
    final_image[220:420,220:420,:] = img4
    final_image[155:275,155:275,:] = center_image
    

    plt.imshow(final_image)
    plt.axis("off")
    plt.show()
    
    return final_image
    
```

```python
img1 = cv2.imread("./../data_sets/collage-filter/ImageCollage_dataset/top_left.jpg")
img2 = cv2.imread("./../data_sets/collage-filter/ImageCollage_dataset/top_right.jpg")
img3 = cv2.imread("./../data_sets/collage-filter/ImageCollage_dataset/bottom_left.jpg")
img4 = cv2.imread("./../data_sets/collage-filter/ImageCollage_dataset/bottom_right.jpg")
img5 = cv2.imread("./../data_sets/collage-filter/ImageCollage_dataset/center.jpeg")
```

```python
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
img4 = cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)
img5 = cv2.cvtColor(img5,cv2.COLOR_BGR2RGB)
```

```python
final_image = collage(img1,img2,img3,img4,img5)
```

```python
r = final_image[:,:,0].flatten()
g = final_image[:,:,1].flatten()
b = final_image[:,:,2].flatten()
```

```python
print(r.shape)
print(g.shape)
print(b.shape)
```

```python
data = np.stack([r,g,b],axis=0)
```

```python
data = data.reshape(-1,3)
```

```python
data.shape
```

```python
data = pd.DataFrame(data,columns=["r","g","b"])
```
